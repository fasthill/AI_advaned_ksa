```python
job_sal = {}
for job in emp['JOB'].unique():
    temp = emp[emp['JOB'] == job]
    sal_mean = temp['SALARY'].mean()
    job_sal[job] = sal_mean
    
index = np.arange(len(job_sal))
plt.bar(index, job_sal.values(), tick_label=list(job_sal.keys()))
plt.xlabel('JOBS')
plt.ylabel('Mean Salary')
plt.legend(['Mean Salary'])
plt.show()
```

```python
job_sal.keys()
```

```python
jobs = emp['JOB'].unique()
means = []
for job in jobs:
    means.append(emp[emp['JOB'] == job]['SALARY'].mean())
    
means
```

```python
plt.bar(np.arange(len(jobs)), means, tick_label = jobs)
plt.grid()
plt.show()
```
