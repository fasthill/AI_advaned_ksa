```python
import numpy as np
import pandas as pd

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import keras
```

```python
np.random.seed(3)
tf.random.set_seed(3)
```

```python
dataset = np.loadtxt('../data/ThoraricSurgery.csv', delimiter=',')
print(type(dataset), dataset.shape, dataset.dtype)
```

```python
X = dataset[:,0:17]
Y = dataset[:,17]
```

```python
model = Sequential()
model.add(Dense(30, input_dim=17, activation='relu')) # input_dim = 입력층. 첫 은닉층이 입력층을 포함.
model.add(Dense(25, activation='relu')) # input_dim = 입력층. 첫 은닉층이 입력층을 포함.
model.add(Dense(20, activation='relu')) # input_dim = 입력층. 첫 은닉층이 입력층을 포함.
model.add(Dense(1, activation='sigmoid')) # 출력층
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X, Y, epochs=200, batch_size=50)
```
