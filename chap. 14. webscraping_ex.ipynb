```python
import requests
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
url = 'https://www.alexa.com/topsites/countries/KR'
w_ranking = requests.get(url)
```

```python
soup = bs(w_ranking.content)
```

```python
soup.select('div.listings.table > div.td.DescriptionCell > p > a')
```

```python
s1= soup.select('div.listings.table a')
```

```python
mlist = [s for s in s1[1:]]
df = pd.DataFrame(mlist, columns =['site'])
df
```
