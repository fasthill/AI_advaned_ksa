```python
excel_data_files = glob.glob('data/담당자별_판매량_*사원.xlsx')
for excel_file in excel_data_files:
    df = pd.read_excel(excel_file)
    
    if df.loc[1, '담당자'] == 'A':
        df['담당자'] = 'Andy'
    elif df.loc[1, '담당자'] == 'B':
        df['담당자'] = 'Becky'
    elif df.loc[1, '담당자'] == 'C':
        df['담당자'] = 'Chris'
        
    new_file_name = re.sub('.xlsx', '2.xlsx', excel_file)
    excel_writer = pd.ExcelWriter(new_file_name, engine='xlsxwriter')
    df.to_excel(excel_writer, index=False, sheet_name='담당자별_판매량')
    excel_writer.save()
```
